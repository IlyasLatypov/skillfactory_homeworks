# Практика для метрик классификации

Реализуем бинарную классификацию записей голосов (мужских и женских) на основе характеристик записей их разговоров.

В предложенном датасете есть ряд свойств, определенных по аудиозаписям:

meanfreq: средняя частота голоса (в кГц)<br />
sd: стандартное отклонение частоты голоса<br />
median: медианная частота (в кГц)<br />
Q25: значение в первом квартиле (в кГц)<br />
Q75: значение в третьем квартиле (в кГц)<br />
IQR: интерквартильный размах (в кГц)<br />
skew: ассиметрия<br />
kurt: эксцесс<br />
sp.ent: спектральная энтропия<br />
sfm: энтропия Винера<br />
mode: мода частоты<br />
centroid: частотный центроид<br />
meanfun: средняя основная частота, измеренная по акустическому сигналу<br />
minfun:  минимальная основная частота, измеренная по акустическому сигналу<br />
maxfun: максимальная основная частота, измеренная в акустическом сигнале<br />
meandom: среднее значение доминирующей частоты, измеренной по акустическому сигналу<br />
mindom: минимум доминирующей частоты, измеренной в акустическом сигнале<br />
maxdom: максимум доминирующей частоты, измеренной в акустическом сигнале<br />
dfrange: диапазон доминантных частот, измеренное на звуковой сигнал<br />
modindx: индекс модуляции голоса

Прежде, чем проводить анализ, можно с помощью визуализации оценить, есть ли различия в частотах голосов мужчин и женщин. Можем построить распределения, например, средних частот:

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("input/7_7.csv", low_memory = False)
```

```python
df.columns
```

```python
import seaborn as sns
```

```python
# Прежде, чем проводить анализ, можно с помощью визуализации оценить, есть ли различия в частотах голосов мужчин и женщин.
# Можем построить распределения, например, средних частот:

sns.distplot(df['meanfreq'])
```

```python
X = df.drop(columns='label')
y = df.label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)

model = LogisticRegression()
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

accuracy_score(y_test, y_pred)
```
