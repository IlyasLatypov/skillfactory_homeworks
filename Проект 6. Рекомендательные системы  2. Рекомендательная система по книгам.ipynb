# РЕКОМЕНДАТЕЛЬНЫЕ СИСТЕМЫ НА ПРАКТИКЕ 

Задача построения и использования рекомендательных систем очень сильно отличается от задачи классификации и анализа данных, так как здесь всегда есть пользователь (User), для которого вы составляете список рекомендаций.

В условиях продакшена первая проблема, с которой приходится сталкиваться — это холодный старт (Cold Start). Что рекомендовать человеку, если он только что к вам пришёл? Не знаете? Всё верно, ведь универсального решения ещё никто не придумал. Давайте посмотрим, как решают данную проблему существующие сервисы рекомендаций. 

Вариант 1. На платформе Netflix рекомендации делаются после того, как пользователя попросят оценить несколько фильмов, чтобы сформировать начальные рекомендации. 

Вариант 2. Просто даём пользователю список самых популярных товаров. На первый взгляд, это отличный вариант, чтобы решить все свои проблемы, но нет. Этот подход показывает самые низкие результаты в долгосрочной перспективе.

Вариант 3. Знаменитое Вместе с этим товаром покупают…, или Content based filtering.

Я предлагаю начать с Content based filtering, то есть давать рекомендации к товару (Item).
