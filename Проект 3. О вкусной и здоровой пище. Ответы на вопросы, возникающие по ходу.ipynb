```python
# Вытаскиваю даты, когда были добавлены комментарии

rx = r'(\d\d/\d\d/\d{4})'

df['Reviews Date 3'] = df['Reviews'].str.findall(rx).apply(','.join)
display(df['Reviews Date 3'])

df['Reviews Date 3'].tolist()
```

```python
# Теперь бы разложить собранные даты комментариев в отдельные колонки
# У функции str.split есть замечательный параметр expand=True, позволяющий сразу сделать разбиение на колонки и получить датафрейм
# Подчитал здесь: http://datalytics.ru/all/kak-v-pandas-razbit-kolonku-na-neskolko-kolonok/

new_df = df['Reviews Date 3'].str.split(',',expand=True)
new_df
```

```python
# Переименую колонки

new_df.columns = ['Reviews Date 1','Reviews Date 2']
new_df
```

```python
# Приведу их к формате datetime (https://medium.com/jbennetcodes/dealing-with-datetimes-like-a-pro-in-pandas-b80d3d808a7f)

new_df['Reviews Date 1'] = pd.to_datetime(new_df['Reviews Date 1'], utc=True)
```

```python
new_df
```

```python
new_df['Reviews Date 2'] = pd.to_datetime(new_df['Reviews Date 2'], format='%m/%d/%Y', errors='ignore', utc=True)
```

```python
# # Новый датафрейм с разбитыми данными из нужно мне колонки 
# new = new_df['Reviews Date 1'].str.split(" ", n = 1, expand = True) 
  
# # Возьму теперь первую часть из нового созданного датафрейма и создам колонку с датой
# new_df['Reviews Date 1'] = new[0]
# new_df['Reviews Date 1']

new_df['Reviews Date 2'] = new_df['Reviews Date 2'].dt.date
new_df['Reviews Date 2']
```

```python
new_df['Reviews Date 1'] = new_df['Reviews Date 1'].dt.date
new_df['Reviews Date 1']
```
