Внимание! Конечно, в каждой компании логирование, мониторинг и Service Discovery устроены совершенно по-разному и зависят от того, какие бизнес-задачи решаются. Некоторые компании используют и самописные решения. В этом блоке мы осветим лишь наиболее популярные сервисы, поэтому постоянно отслеживайте новинки индустрии, чтобы держать руку на пульсе.

Мониторинг микросервисов — задача непростая: в режиме реального времени нужно отслеживать состояние как отдельных компонентов, так и системы в целом. Задача усложняется, если помимо технических нужно следить ещё и за значимыми для бизнеса показателями.

Если посмотреть на микросервисную архитектуру со стороны, то каждый микросервис по сути живет своей жизнью, и зачастую поддерживается разными командами. Поэтому при организации их взаимодействия что-то обязательно пойдёт не так.

Конечно, каждый сервис в случае ошибки будет отправлять вызывающему сервису подробности произошедшего, а тот, в свою очередь, завернёт это в ответ с указанием причины. Однако, чтобы держать руку на пульсе, нам необходимо постоянно отслеживать все цепочки взаимодействия. 

В идеальном варианте логи сообщения каждый сервис будет складировать независимо, а потом логи будут собираться вместе специальным агрегатором.

КАК ЭТО СДЕЛАТЬ?

Итак, самый известный вариант — это стек ELK: Elasticsearch, Logstash и Kibana. 

Logstash — это инструмент для сбора данных с открытым исходным кодом с возможностями конвейерной обработки данных в реальном времени. Инструмент позволяет динамически идентифицировать данные из различных источников и нормализовать их, с помощью выбранных фильтров.
Elasticsearch — это не просто база данных NoSQL, но и полноценный движок полнотекстового поиска и аналитики. Написан на языке Java.
Kibana — это панель визуализации данных. Данные представляются в виде различных диаграмм.
Все инструменты работают в связке: 

Logstash отвечает за агрегацию логов для каждого микросервиса → поставляет входящий поток данных в Elasticsearch для хранения, классификации и поиска → Kibana получает доступ к данным Elasticsearch для их визуализации в различных визуальных форматах, например в виде дашборда.

Помимо стека ELK существует связка Prometheus и Grafana. 

Prometheus выступает в качестве системы агрегации логов и базы данных временных рядов для хранения метрик, а Grafana помогает построить дашборд для визуализации данных.

ПРОБЛЕМА

В случае микросервисного подхода к разработке остро встаёт необходимость взаимного обнаружения связанных компонентов системы. 

Представим на секунду ситуацию: у нас десятки тысяч сервисов, раскиданных по тысячам серверов. Как сервисы будут находить друг друга? Как они будут определять состояние другого сервера?

Конечно, нам бы хотелось решить эту задачу автоматически. Например, созданием такого журнала, в который сервисы бы писали своё состояние и адреса.

Хорошая новость: всё уже придумано за нас! Для решения этой задачи используются протоколы обнаружения сервисов (Service Discovery). В дополнение к ним существует ряд инструментов, например:

- Consul: http://www.consul.io/
- Etcd: https://etcd.io/
- ZooKeeper: http://zookeeper.apache.org/

На текущий момент Consul от компании HashiCorp является одним из самых популярных в современной разработке. Этот сервис позволяет не только обнаружить сервисы, но и конфигурировать их.
