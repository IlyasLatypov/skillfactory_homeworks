При создании микросервисной архитектуры требуется организация обмена данными между сервисами.

Давайте разберёмся, какие существуют типы организации общения:

Синхронный

Один сервис обращается к другому и ожидает ответа.

Для организации синхронного взаимодействия используется протокол HTTP или HTTPS. Разработка и отладка просты, однако сервис должен быть постоянно доступен, в противном случае обмен сообщениями остановится. 

⇢ Подробнее мы поговорим про HTTP в следующем блоке.

Асинхронный

Сервисы взаимодействуют между собой путем передачи сообщений. Таким образом, сервис не ожидает ответ, а продолжает работу. Нужный сервис принимает сообщение и начинает его обработку.

Для организации асинхронного взаимодействия используются очереди сообщений. ⇣ О них мы поговорим ниже.

БРОКЕРЫ СООБЩЕНИЙ

Брокеры сообщений (ещё их называют диспетчеры очереди) помогают организовать межсистемный обмен сообщениями и являются связующим звеном между различными процессами в приложениях. Как Почта России, но быстрее и надежнее!

Как правило, в таких системах используется паттерн проектирования издатель → очередь → подписчик (producer → queue → consumer).

Благодаря брокерам сообщений отправителю не нужно ничего знать о потребителе или потребителях, поэтому связь между сервисами значительно упрощается. Однако поддержка инфраструктуры с участием очереди становится немного сложнее.

Чтобы обеспечить хранение сообщений в очереди до момента прочтения их получателем, доступ к очереди должен быть асинхронным. 

Структура данных очереди представлена следующими правилами:

- первым «читается» сообщение, пришедшее первым;
- новые элементы добавляются только в конец очереди;
- чтение и удаление происходит только с начала очереди.

Кроме того, формат сообщений должен быть определён (например XML или JSON). Однако это сильно зависит от типа реализации очереди — бывают случаи, когда определять формат нет необходимости.

Наиболее популярными являются следующие программные реализации очереди:

- RabbitMQ
- Apache Kafka
- ActiveMQ

Нужно также отметить, что с помощью протокола HTTP возможно организовать не только синхронное, но и асинхронное взаимодействие, однако это выходит за рамки нашего курса.
