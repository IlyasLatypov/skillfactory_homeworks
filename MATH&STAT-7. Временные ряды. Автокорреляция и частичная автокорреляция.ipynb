АВТОКОРРЕЛЯЦИЯ (ACF)

Идея использования авторегрессионных моделей состоит в том, что значения ряда зависят от исторических данных — значений в предыдущие моменты времени.

Причём, чем дальше в историю мы уходим, тем слабее эта зависимость.

Математически эту связь можно выразить при помощи автокорреляционной функции:
