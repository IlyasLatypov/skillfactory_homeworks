```python
first_line = [x*y for x in range(2, 100, 6) for y in range (7, 1, -2)]
second_line = [x ** 0.5 for x in range(1000, 1101, 2)]
third_line = [x**2 for x in range(51)]

big_secret = np.array([first_line, second_line, third_line, second_line, first_line])
```

```python
big_secret
```

```python
"""
Чему равна сумма элементов последнего столбца массива? Ответ округлите до двух цифр после запятой:
"""

result = big_secret[:5, -1:].sum()
round(result, 2)
```

```python
"""
Выделите из каждой строки массива big_secret первые 5 элементов.

Чему равна сумма элементов главной диагонали получившейся матрицы? Округлите ответ до двух цифр после запятой:

"""

big_secret_1 = big_secret[:5, :5]

# сумма элементов главной диагонали
sumMain = 0

for i in range(5):
    # У элементов главной диагонали совпадают оба индекса.
    # Из матрицы извлекается элемент и добавляется к сумме.
    sumMain += big_secret_1[i][i]
    
# вывод сумм
print(round(sumMain, 2))
```

```python
"""
Выделите из каждой строки массива big_secret последние 5 элементов. Чему равно произведение элементов главной диагонали получившейся матрицы?
Введите полученный результат без изменений и округлений.

"""

big_secret_2 = big_secret[:5, -5:]

# сумма элементов главной диагонали
sumMain2 = 1

for i in range(5):
    # У элементов главной диагонали совпадают оба индекса.
    # Из матрицы извлекается элемент и добавляется к сумме.
    sumMain2 *= big_secret_2[i][i]
    
# вывод сумм
print(round(sumMain2, 2))
```

```python
"""
Выделите из каждой строки обновлённого массива big_secret последние 5 элементов.
Чему равно произведение элементов главной диагонали получившейся матрицы?
Введите полученный результат без изменений и округлений.

"""

big_secret_4 = big_secret[:5, -5:]

display(big_secret)

# сумма элементов главной диагонали
sumMain4 = 1

for i in range(5):
    # У элементов главной диагонали совпадают оба индекса.
    # Из матрицы извлекается элемент и добавляется к сумме.
    sumMain4 *= big_secret_4[i][i]
    
# вывод сумм
print(sumMain4)

np.diagonal(big_secret_4).prod()
```
