# Визуализация процесса обучения

Пример графика, который показывает ошибку модели в зависимости от времени — c течением времени обучения падает ошибка модели. Это то, к чему нужно стремиться. 

Визуализация — полезный инструмент для отслеживания проблем в процессе обучения. Например, если график ошибки на валидации становится выше, чем график ошибки на обучении, то можно сказать, что модель переобучилась. 

В дальнейшем для визуализации вы можете использовать такие инструменты, как TensorBoard от Google. Это бесплатный инструмент, позволяющий в реальном времени смотреть на метрики обучения, показывать графики ошибок. С его помощью вы сможете сравнивать графики по разным моделям, чтобы получить представление о том, какая модель лучше или хуже справилась, и какие ошибки возникли на пути.
