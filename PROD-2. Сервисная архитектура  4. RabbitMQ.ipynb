Какую бы IDE вы ни выбрали, суть работы в них всегда одинаковая — создаёте проект, добавляете туда новый файл .py и… приступаете к работе. 

→ А таблицы горячих клавиш сделают жизнь ещё более приятной! Список горячих клавиш для PyCharm.

Как вы помните, объект из памяти одного процесса нельзя напрямую перенести в память другого, ведь он может иметь сложную структуру и быть разложен в оперативной памяти по кусочкам. Чтобы сохранить целостность объекта, и применяется сериализация.

При сохранении и загрузке моделей мы проводим эти операции лишь один раз, поэтому это не увеличивает нагрузку на процессы. Если мы хотим организовать взаимодействие микросервисов, необходимо, чтобы наши сервисы могли общаться между собой. Для этого нужно при отправке сериализовывать каждое сообщение (например, вектор входных данных), десериализовывать его после получения в сервисе для подачи в модель и уже после преобразовывать в поток битов предсказания. А это требует ресурсов!

Совет. Если добавить сжатие данных, то потребуется больше ресурсов CPU на их чтение и запись, зато сам объём данных будет меньше. Таким образом, мы уменьшим нагрузку на сеть и диск.

Для взаимодействия между сервисами чаще всего используется уже знакомый вам формат JSON. 

⇢ Для работы с этим форматом используется библиотека json, которая очень похожа на pickle. 

Для полноценного понимания принципов работы с очередями сообщений и сериализацией, мы разберём диспетчер сообщений RabbitMQ. Почему именно RabbitMQ? Он прост в использовании, максимально эффективен, работает на разных системах и, видимо, по этим причинам пользуется большой популярностью.

Для примера работы с этой очередью мы возьмем интеграцию RabbitMQ и Docker. 

⇢ С Docker вы познакомитесь уже в следующем модуле. Пока лишь скажем, что Docker — это платформа для разработки и доставки приложений.

Давайте установим все инструменты, которые нам понадобятся.

УСТАНОВКА ИНСТРУМЕНТОВ

→ Шаг 1. Установим Docker. Чтобы установить Docker, пройдите по ссылке. 

→ Шаг 2. Установим RabbitMQ. Теперь, когда Docker уже есть в нашей системе, достаточно просто написать одну bash-команду в командной строке: 

docker run -it --rm --name rabbitmq -p 5672:5672 -p 15672:15672 rabbitmq:3-management

Очередь запущена и работает. К ней можно обратиться по адресу localhost:5672.

→ Шаг 3. Подключим библиотеки. RabbitMQ использует протокол AMQP. Чтобы с ним работать на Python, нам понадобится pika. Установим её через pip команду.

pip install pika --upgrade

Итак, всё готово к работе.
