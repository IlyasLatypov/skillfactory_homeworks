```python
# Начальный набор данных по распределениям вероятностей, записанных в виде словаря

x = {'1': 0.2, '2': 0.1, '3': 0.4, '4': 0.3}
x

y = {'-1': 0.5, '0': 0.2, '1': 0.1, '2': 0.2}
y
```

```python
# Математическое ожидание x (оно же среднее)

expectation_x = 0

for k, v in x.items():
    expectation_x += int(k) * v
    
print(expectation_x)
```

```python
# Дисперсия x

import numpy as np 

var_x = 0

for k, v in x.items():
    var_x += (np.square(int(k) - expectation_x)) * v 
    
print(round(var_x, 2))
```

```python
# Стандартное отклонение x

σ_x = np.sqrt(var_x)
print(round(σ_x, 2))
```

```python
# Вес одного гвоздя 10 граммов, вес пустой упаковки 20 граммов.
# На основании ответов, полученных в первой части, найдите средний вес упаковки с гвоздями и его стандартное отклонение.

item_weight = 10
package_weight = 20
x_new = {}

for k, v in x.items():
    x_new.update({(int(k) * item_weight) + package_weight: v})
    
x_new
```

```python
# Математическое ожидание (оно же среднее)

expectation_x_new = 0

for k, v in x_new.items():
    expectation_x_new += int(k) * v
    
print(expectation_x_new)
```

```python
# Дисперсия

import numpy as np 

var_x_new = 0

for k, v in x_new.items():
    var_x_new += (np.square(int(k) - expectation_x_new)) * v 
    
print(round(var_x_new, 2))
```

```python
# Стандартное отклонение

σ_x_new = np.sqrt(var_x_new)
print(round(σ_x_new, 1))
```

```python
# Задание 2.11.1

"""
Случайный покупатель в книжном магазине купит одну, две или три книги с вероятностью 0.3, 0.4 и 0.3, соответственно.

Каждая книга стоит 400 рублей. Найдите математическое ожидание расходов покупателя магазина.

"""

task_2_11_1 = {'1': 0.3, '2': 0.4, '3': 0.3}


book_cost = 400

task_2_11_1_400 = {}

for k, v in task_2_11_1.items():
    task_2_11_1_400.update({(int(k) * book_cost): v})

task_2_11_1_400

# Математическое ожидание task_2_11_1 (оно же среднее)

expectation_task_2_11_1_400 = 0

for k, v in task_2_11_1_400.items():
    expectation_task_2_11_1_400 += int(k) * v
    
print(expectation_task_2_11_1_400)
```

```python
# Задание 2.11.4

"""
Оценки студентов на международной онлайн олимпиаде распределены нормально со средним 700 баллов
и стандартным отклонением 120 баллов.

Диплом первой категории выдается участникам, набравшим более 820 баллов.

Диплом второй категории — участникам, набравшим от 730 до 820 баллов.

Для решения задачи используйте функции norm.cdf() и norm.ppf() из пакета scipy.stats.

"""
μ = 700
X = 120

# P(X > 820)

round((1 - norm.cdf(820, loc = 700, scale = 120)), 2)

# P(820 > X > 730)
# Вероятность значений в определенном интервале можно найти как разность вероятностей:
# (P(a≤X≤b)=P(X≤b)-P(X≤a)), каждую из которых можно вычислить, используя norm.cdf().

a = norm.cdf(820, loc = 700, scale = 120)
b = norm.cdf(730, loc = 700, scale = 120)
print(round((a - b), 2))

round(norm.ppf(0.05, loc = 700, scale = 120), 0)
```
