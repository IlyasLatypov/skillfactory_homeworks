КАКОЙ КЕЙС РЕШАЕМ?

Вы работаете в крупной компании, занимающейся онлайн-ритейлом (розничными продажами в интернете).

Ваш работодатель хочет увеличить средний чек продаж. Вы знаете, что чем точнее рекомендовать нужный товар пользователю (в нашем случае, товар с более высоким средним чеком), тем выше вероятность покупки этого товара. Простая логика. Мы нашли золото!

У вас будет история оценок пользователя вместе с его отзывом на товар. Вы можете использовать текст рецензии в качестве дополнительной информации. Все оценки пользователей нормированы для бинарной классификации: если человек поставил оценку продукту 4 и больше, то мы считаем, что продукт ему понравился, если меньше 4, то продукт не понравился.

Чтобы определить победителя в соревновании, мы будем использовать знакомый вам Kaggle. Подробное описание соревнования и датасета доступно по клику:

МЕТРИКА

Команда-победитель определяется по наибольшему значению RocAuc.

RocAuc

В этом соревновании будет использоваться метрика RocAuc. Это одна из самых популярных метрик в задачах бинарной классификации. Почему мы используем бинарную классификацию? Потому что мы хотим предсказывать, понравится ли товар пользователю или нет.

RocAuc чаще всего воспринимают как площадь под кривой. Подробнее вы можете прочитать здесь. Вы должны понимать, что эта метрика зависит не от абсолютных значений ваших предсказаний, а от относительных: если вы умножите или увеличите все ваши предсказания на 10, то RocAuc не изменится.

RocAuc — это ещё и симметричная метрика: вы можете получить 0.3, а потом перевернуть ваши предсказания и получить 0.7.  Если вы будете делать случайные предсказания, то получите 0.5. Таким образом, случайный классификатор в нашей задаче всегда будет выдавать оценку, близкую к 0.5. Чем выше значения RocAuc, тем лучше. Вам нужно добиться максимального значения.
