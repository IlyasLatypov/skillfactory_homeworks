```python
"""
Какой средний балл, полученный мальчиками на экзамене по чтению? Ответ округлите до трёх цифр после запятой.

"""

import csv

f = csv.reader(open('StudentsPerformance.csv'))

count_students_reading_score = 0
males = 0

next(f)
for line in f:
    if line[0] == 'male':
        count_students_reading_score += int(line[6])
        males += 1
            
avg_count_students_reading_score = (count_students_reading_score / males)

round(avg_count_students_reading_score, 3)
```

```python
f = open('StudentsPerformance.csv')

max_score = 0
read = []

for line in f:
    info = line.split(',')
    gender = info[0][1:-1] #отмечаем, чтобы потом проверить, что первая строка с заголовками не считается
    math_score = info[5][1:-1] #ищем в списке балл по математике
    read_score = info[6][1:-1] #то же самое с баллом по чтению
    if gender != 'gender':
        if int(math_score) > int(max_score):
            max_score = math_score #определяем максимальный бал по математике.
            
for line in f:
    info = line.split(',')
    gender = info[0][1:-1] #отмечаем, чтобы потом проверить, что первая строка с заголовками не считается
    math_score = info[5][1:-1] #ищем в списке балл по математике
    read_score = info[6][1:-1] #то же самое с баллом по чтению
    if gender != 'gender':
        if int(math_score) == max_score:
            read.append(int(read_score)) #добавляем в список все оценки по чтению тех, у кого максимальный средний балл по математике
            
            
f.close()   

print(max_score)
print(read)
```
