# ЗАЧЕМ НУЖНА СТАТИСТИКА?

В этом модуле мы переходим от теории вероятностей к статистике.

У нас уже есть одна наука, работающая со случайностями. Зачем же ещё одна? Приведём пример.

ПРИМЕР

В идеальной модели монетка падает орлом и решкой с вероятностью 50 %.
→ Теория вероятностей подскажет, с какой вероятностью выпадет 2 орла подряд.

Пусть монетку бросили 5 раз, выпало 4 орла и 1 решка.
→ Статистика ответит на вопрос: насколько правдоподобно, что монетка идеальная?

Теория вероятностей не спрашивает, подходит ли модель к реальности, а только прогнозирует будущее на основе модели (возможно, неверной).

Статистика проверяет, насколько модель подходит к реальности, исходя из того насколько сильно идеальные прогнозы модели не соответствуют наблюдаемым реальным данным.

## ЧТО БУДЕМ ИЗУЧАТЬ?

В этом модуле мы:

познакомимся с понятиями генеральной совокупности и выборки;
узнаем, чем отличаются параметры от их оценок и какие оценки являются хорошими;
поговорим о том, где эти хорошие оценки можно брать.
Оказывается, в стандартных случаях есть готовые рецепты:

выборочная пропорция;
выборочное среднее;
выборочная дисперсия;
стандартное отклонение;
выборочная медиана;
выборочная корреляция.
Также нам важно понять, как отличить закономерность от случайности: 

проверить, действительно ли эффективно новое лекарство;
оценить, отличается ли привлекательность двух рекламных роликов; 
изучить, существуют ли действительно отличия в освоении технических наук между мужчинами и женщинами.
Для того, чтобы проверять наши предположения, мы разберем принцип тестирования гипотез и познакомимся с понятием доверительных интервалов.

**Определение. Генеральная совокупность — это вся исследуемая популяция.**

В случае, если всю генеральную совокупность исследовать не удаётся, берут её часть — выборочную совокупность (или выборку), исследуют её, а потом выводы обобщают на генеральную совокупность.

**Определение.  Выборка — часть исследуемой популяции, доступная для наблюдений.**

Например, генеральная совокупность — это все жители вашего дома, а выборка — те из них, кто пришёл на собрание по поводу установки шлагбаума.

### ГЛАВНАЯ ЦЕЛЬ СТАТИСТИКИ — ПО НАБЛЮДЕНИЯМ ИЗ ВЫБОРКИ СДЕЛАТЬ ВЫВОДЫ О ПОПУЛЯЦИИ.

### ВАЖНЫМ УСЛОВИЕМ ПОЛУЧЕНИЯ ХОРОШИХ ОЦЕНОК ЯВЛЯЕТСЯ СЛУЧАЙНОСТЬ ВЫБОРКИ.

Определение. На формальном языке выборкой мы будем называть набор независимых одинаково распределённых случайных величин x1, x2, x3 и тд. Их количество называется размером выборки.

Здесь возникает проблема репрезентативности. Выборка должна обязательно по характеристикам соответствовать генеральной совокупности. Например, если в нашей генеральной совокупности 70 % женщин и 30 % мужчин, то и в выборке должно быть 70 % женщин и 30 % мужчин. Или, если в генеральной совокупности 90 % людей с высшим образованием, то и в выборке должен быть такой же процент.

### ИТАК, ВЫБОРКА ОБЯЗАТЕЛЬНО ДОЛЖНА БЫТЬ РЕПРЕЗЕНТАТИВНОЙ.

Определение. Репрезентативная выборка — это выборка, обладающая всеми свойствами исходной популяции, значимыми с точки зрения задач исследования.

### КАКИЕ ОЦЕНКИ СЧИТАТЬ ХОРОШИМИ?

Мы уже поняли, что выборка обязательно должна быть случайной. Как теперь определить, какие оценки лучше, а какие хуже?

У оценок есть два ключевых критерия качества:

несмещённость;
состоятельность.

Несмещённость означает, что в среднем мы оценим параметр верно.

Состоятельность означает, что при увеличении размера выборки ошибка оценки уменьшается.
