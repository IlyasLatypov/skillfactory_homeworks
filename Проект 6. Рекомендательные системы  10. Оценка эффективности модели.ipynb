Итак, наша рекомендательная система работает. Возможно, мы даже внедрили её в продакшн. На практике вам всегда нужно будет понимать, как решение повлияет на результаты бизнеса, и, что немаловажно, уметь доносить это до коллег. 

Выделяют пять групп бизнес-показателей, на которые рекомендательные системы оказывают самое существенное влияние: 

1. СTR (Кликабельность). Сколько раз пользователи кликнули на рекомендацию.
2. Применение и конверсия. Как пользователи взаимодействуют с продуктом из списка рекомендаций. Например, сколько пользователей посмотрело рекомендованный фильм?
3. Продажи и выручка. Как увеличились продажи после внедрения алгоритма? 
4. Охват различных категорий товаров. Если пользователь хочет купить деловой костюм, то мы можем порекомендовать в дополнение галстук (аксессуары) и ботинки (обувь). Как в таком случае изменятся продажи товаров из других категорий? 
5. Поведение пользователей: churn rate (отток клиентов), retention rate (удержание клиентов), длительность сессии и другие.

По данным Netflix, 75 % контента пользователи выбирают на основе рекомендаций. Компании также удалось снизить отток клиентов на несколько процентов, что, в конечном итоге, приносит около миллиарда долларов в год (компания использует модель подписки). В свою очередь YouTube сообщает, что 60 % кликов на домашней странице приходятся на блок с рекомендациями.

С другой стороны, выгода от внедрения рекомендательных систем во многих случаях может быть не такой очевидной — следует понимать, на какие метрики смотреть. Так, например, после внедрения рекомендательных систем в онлайнмагазине, было замечено, что рост по рекомендуемым товарам составил всего 0,3 %. При этом средний рост по товарам в той же категории достиг 26 %. Получается, что внедрение рекомендаций может стимулировать дополнительные продажи, несмотря на то, что пользователь не выбирал продукт из рекомендуемого списка.
