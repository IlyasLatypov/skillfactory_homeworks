### Парные дискретные распределения

Часто бывает, что одной случайной величины для описания вероятностной модели задачи недостаточно. Например, один покупатель купил у нас \(X\) эклеров, а второй — \(Y\). Как смоделировать распределение суммарной покупки? Для этого как раз и нужны многомерные и в частности парные распределения.

Многомерные случайные величины удобно рассматривать как случайные векторы, где каждая координата — обычная случайная величина. Чтобы задать двумерный случайный вектор \((X, Y)\), нужно определить множество пар значений \((a, b)\). И для каждой пары указать вероятность \(P(X=a, Y=b)\).

Частное распределение
В ходе решения задачи может понадобится перейти от многомерного распределения к набору одномерных. Это может быть нужно для выяснения независимости или нахождения матожидания и дисперсии каждой координаты в отдельности.

Определение:  Marginal (Частное) распределение — распределение конкретной координаты случайного вектора.

### Ковариация и корреляция

Ковариация и корреляция намного больше интересуют нас с точки зрения статистики, нежели с точки зрения теории вероятностей. Однако, чтобы было понятно, о чем далее в статистике пойдёт речь, давайте дадим определение.

Напомним, что дисперсия случайной величины вычисляется как разность между матожиданием ее квадрата и квадратом ее матожидания

\(VX=E(X^2)-(EX)^2\)
\(VY=E(Y^2)-(EY)^2\)

Можно переписать эти формулы немного в другом виде, заменив квадраты на непосредственное умножение. Дисперсия \(Х\) станет разностью между произведением матожидания \(Х\) с самим собой и произведением двух матожиданий \(Х\)

\(VX=E(X^2)-(EX)^2\)
\(VX=E(X \cdot X)-EX \cdot EX\)
\(VY=E(Y^2)-(EY)^2\)
\(VY=E(Y \cdot Y)-EY \cdot EY\)

Можно сказать, что дисперсия является вариацией одной случайной величины.

Определение:  Ковариация же является совместной вариацией двух случайных величин

Определение:  Корреляцией в теории вероятностей называется отношение ковариации к произведению стандартных отклонений
