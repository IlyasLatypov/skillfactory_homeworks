Прочитав множество статей про использование рекомендательных систем на практике, я понял, что руководства по применению не существует. Чаще всего встречались объемные тексты, смысл которых можно было передать одной фразой: «Production — это хорошо, а ваша модель должна быть хорошей и быстрой» (Ну-ну…). Тогда я и решил, что мне нужно разобраться во всём самому и протестировать различные варианты.

✔ Онлайн. Обучаем модель, используем её предсказания на новых или старых пользователях. 

Плюсы: Точность, возможность использования с новыми пользователями.

Минусы: Затраты на предсказание и обучение.

✔ Офлайн. Один раз обучаем модель и... всё! А что значит «один раз обучаем модель»? Всё просто: мы обучили модель, сделали векторизованные представления наших объектов (embeddings) и пользуемся ими в процессе.

Плюсы: Нет затрат на предсказание во время работы (inference). Можно хранить все рекомендации в базе данных, останутся только затраты на обращение к ней.

Минусы: Не можем давать пользователям персонализированных рекомендаций.

Я решил остановиться на офлайн-модели, так как онлайн-модель нельзя использовать на новых пользователях.
