Требования продакшн-среды несомненно влияют и на выбор сервисной архитектуры.

Существует два основных типа: монолитная и микросервисная архитектуры.

Монолитная

Все контексты приложения запускаются на сервере с помощью внутрипроцессных взаимодействий. То есть приложение включает в себя все функции, и они взаимодействуют в одной запущенной программе.

Микросервисная

Подход, позволяющий инкапсулировать определённые контексты приложения. Таким образом, каждая функция вынесена в отдельную небольшую программу или сервис.

ПРЕИМУЩЕСТВА МИКРОСЕРВИСНОГО ПОДХОДА В РАЗРАБОТКЕ

✔ Мы повышаем автономность различных частей продукта. Мы можем разрабатывать их раздельно и разными командами (отслеживая при этом версионность!)

✔ Мы обеспечиваем гибкую масштабируемость. За счет модульности приложения, мы можем масштабировать только ту часть, которая требует дополнительных ресурсов.

✔ Мы повышаем стабильность. Если что-то упало, то будет недоступен только один функционал, а не все приложение. 

✔ Наконец, микросервисы позволяют запускать приложение на разных серверах, что обеспечивает горизонтальную масштабируемость.

Давайте разберёмся в этом подробнее. При создании действительно крупных приложений существует два подхода: 

Вертикальное масштабирование (его ещё называют scale up): заменить сервер сервером с большей вместимостью и мощностью. Как вы понимаете, у всего есть предел.

Горизонтальное масштабирование (или scale out): добавить ещё один сервер и объединить сервера в кластер.

НЕДОСТАТКИ МИКРОСЕРВИСНОЙ АРХИТЕКТУРЫ

Несмотря на обилие плюсов от использования микросервисной архитектуры, таких как модульность и независимость реализаций, в ней присутствуют и недостатки:

- Сложность разработки из-за наличия различных технологий
- Накладные расходы на сериализацию и десериализацию сообщений
- Время работы напрямую зависит от того, как сервисы общаются между собой.

На текущий день выбор монолитной или микросервисной архитектуры должен быть основан на понимании всех требований к разрабатываемому приложению.
